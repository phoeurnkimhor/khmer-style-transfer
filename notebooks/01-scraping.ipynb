```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:146.0) Gecko/20100101 Firefox/146.0'}
url = 'https://km.wikipedia.org/wiki/%E1%9E%94%E1%9F%92%E1%9E%9A%E1%9E%87%E1%9E%BB%E1%9F%86%E1%9E%9A%E1%9E%BF%E1%9E%84%E1%9E%96%E1%9F%92%E1%9E%9A%E1%9F%81%E1%9E%84%E1%9E%81%E1%9F%92%E1%9E%98%E1%9F%82%E1%9E%9A'

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

links = soup.select('div li a')
base_url = "https://km.wikipedia.org"

stories = [
    {'title': a.get_text(strip=True), 'url': urljoin(base_url, a['href'])}
    for a in links
    if a.has_attr('href') and a['href'].startswith('/wiki/%')
]
```

```python
import pandas as pd
import time
data = []

for i in range(len(stories)):

    response = requests.get(stories[i]['url'], headers=headers)
    # print(response.text)
    soup = BeautifulSoup(response.text, "html.parser")
    paragraphs = soup.select('div.mw-parser-output > p')
    khmer_text = "\n\n".join([p.get_text(strip=True) for p in paragraphs if p.get_text(strip=True)])
    data.append({'title' : stories[i]['title'], 'text' : khmer_text})
    time.sleep(1)

df = pd.DataFrame(data)
df
```

```python
index_str = "0-13 15 19 20 21 31-33 35-38 43 44 46 49 50-55 58-59 60 61 63 65-68 72-76 78-89 91-96"

# Convert to a list of indices
indices_to_drop = []
for part in index_str.split():
    if '-' in part:
        start, end = map(int, part.split('-'))
        indices_to_drop.extend(range(start, end + 1))
    else:
        indices_to_drop.append(int(part))

# Drop rows by index
df = df.drop(indices_to_drop).reset_index(drop=True)
```
